### Pulizia DF

```python
# STEP 1: Convertiamo i valori in stringa
df['discounted_price'] = df['discounted_price'].astype(str)
df['actual_price'] = df['actual_price'].astype(str)

# STEP 2: Rimuoviamo il simbolo ₹, virgole e spazi
df['discounted_price'] = df['discounted_price'].str.replace('₹', '', regex=False).str.replace(',', '').str.strip()
df['actual_price'] = df['actual_price'].str.replace('₹', '', regex=False).str.replace(',', '').str.strip()

# STEP 3: Convertiamo in float
df['discounted_price'] = df['discounted_price'].astype(float)
df['actual_price'] = df['actual_price'].astype(float)
```

```python
# rating: da stringa a float (gestisce anche eventuali valori non convertibili)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# rating_count: rimuove le virgole e converte in float
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '').astype(float)

# discount_percentage: rimuove il simbolo % e converte in float
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '').astype(int)
```
