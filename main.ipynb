### Pulizia DF

```python
# STEP 1: Convertiamo i valori in stringa
df['discounted_price'] = df['discounted_price'].astype(str)
df['actual_price'] = df['actual_price'].astype(str)

# STEP 2: Rimuoviamo il simbolo ₹, virgole e spazi
df['discounted_price'] = df['discounted_price'].str.replace('₹', '', regex=False).str.replace(',', '').str.strip()
df['actual_price'] = df['actual_price'].str.replace('₹', '', regex=False).str.replace(',', '').str.strip()

# STEP 3: Convertiamo in float
df['discounted_price'] = df['discounted_price'].astype(float)
df['actual_price'] = df['actual_price'].astype(float)
```

```python
# rating: da stringa a float (gestisce anche eventuali valori non convertibili)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# rating_count: rimuove le virgole e converte in float
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '').astype(float)

# discount_percentage: rimuove il simbolo % e converte in float
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '').astype(int)
```

```python
# verifichiamo se actual_price ha outliers

# Calcoliamo i quartili Q1 (25%) e Q3 (75%)
Q1 = df['actual_price'].quantile(0.25)
Q3 = df['actual_price'].quantile(0.75)

# Calcoliamo l'intervallo interquartile (IQR)
IQR = Q3 - Q1

# Definiamo i limiti oltre i quali consideriamo un valore come outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Verifichiamo quanti valori sono outlier
outliers = df[(df['actual_price'] < lower_bound) | (df['actual_price'] > upper_bound)]

# Stampiamo il numero di outlier e mostriamo i primi 5
print(f"Numero di outlier in actual_price: {len(outliers)}")
print(outliers[['product_name', 'actual_price']].head())


# Visualizziamo i dati con un boxplot
plt.figure(figsize=(10, 1))
sns.boxplot(x=df['actual_price'])
plt.title("Boxplot di actual_price")
plt.show()
```

```python
## segmentazione dei dati per fascia di prezzo

# Creiamo una nuova colonna 'price_range' per categorizzare i prezzi
df['price_range'] = 'Basso'  # inizializziamo tutto come 'Basso'

# Applichiamo le regole per le fasce di prezzo
df.loc[(df['actual_price'] >= 1000) & (df['actual_price'] <= 5000), 'price_range'] = 'Medio'
df.loc[(df['actual_price'] > 5000) & (df['actual_price'] <= 15000), 'price_range'] = 'Alto'
df.loc[df['actual_price'] > 15000, 'price_range'] = 'Premium'
```
